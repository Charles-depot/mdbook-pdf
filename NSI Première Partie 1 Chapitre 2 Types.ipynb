1. Exécuter les cellules ci-dessous :

2. L'instruction <code>==</code> permet de réaliser un test d'égalité. Elle renvoie donc le booléen <code>True</code> si les deux membres sont égaux et le booléen <code>False</code> sinon.<br>Exécuter les cellules ci-dessous :

````python
0.25**2 == 1/16
````

````python
0.7**2 == 0.49
````

````python
2*0.1 == 0.2
````

````python
3*0.1 == 0.3
````

<p style="color:rgb(255,0,0)">A retenir</p>
<div class="alert alert-danger" style="border-left: 15px solid #a94442;border-radius:2px;">
Les calculs algébriques avec les nombres entiers (type <code>int</code>) sont tous exacts.<br><br>
En revanche, il n'en est pas de même avec les nombres flottants (type <code>float</code>) : certains calculs sont exacts et d'autres non. Enfin, c'est ce qu'il semble avec les exemples ci-dessus. Ces "erreurs" sont dues à la façon dont les nombres flottants sont représentés en binaire dans l'ordinateur. En fait, ces erreurs sont contrôlables et les calculs avec les flottants sont assurés être exacts avec 16 décimales.<br><br>
Ce qu'il faut retenir ici, c'est qu'il ne faut pas faire des tests d'égalités brutaux entre deux nombres flottants. En fait, il faut comparer l'écart entre $1$ et le quotient des deux valeurs que l'on veut tester et vérifier si cet écart est plus petit qu'un nombre prédéfini appelé $\varepsilon$-machine.<br>
Cette valeur permet d'assurer que les calculs avec les nombres flottants sont rigoureux. Elle vaut $2^{-23}$ pour une architecture $32$ bits et $2^{-52}$ pour une architecture $64$ bits.<br>Ainsi, pour tester si <code>3*0.1</code> est bien égal à <code>0.3</code>, on utilise le test suivant (Capytale utilise une architecture $32$ bits) :
    
```python
    abs(1-3*0.1/0.3) < 2**(-23)
```
La valeur renvoyé est bien <code>True</code>.<br><br>
