<h2 class="alert alert-info">Exercice 1 : édition et exécution d'une cellule de code</h2>

Un notebook jupyter consiste en **cellules** individuelles pouvant contenir du **code** Python, du **texte** (avec formatage Markdown, HTML et LaTeX) ou des **graphiques**. Commençons par les cellules de type Code.

1. Placez le curseur dans la 1ère cellule <code>Entrée[ ]:</code> ci-dessous puis frappez `<enter>`. Que se passe-t-il ? Rien d'autre que l'insertion d'un saut de ligne... C'est normal, car dans une cellule de type "Code" (c'est-à-dire préfixée par <code>Entrée[ ]:</code>) on est par défaut en mode **édition de code** Python !


Pour **exécuter le code** Python d'une cellule, il faut en effet frapper `<maj-entrée>` (ou cliquer sur le bouton <button class='fa fa-step-forward icon-step-forward'></button>, ou encore utiliser le menu `Cellule > Exécuter les cellules et sélectionner la suivante`). Essayez !

```python
15 + 24
```

```python
'Hello tout le monde !'
```

Une ligne <code>Sortie[ ]:</code> devrait être apparue en dessous de la 1ère expression, affichant le résultat de celle-ci ! Le curseur est alors automatiquement passé dans la cellule d'en-dessous que vous pouvez exécuter à son tour.

_Notez qu'il est aussi possible de frapper `<ctrl-entrée>` (identique à `Cellule > Exécuter les cellules `). Dans ce cas, le curseur **reste dans la cellule** après l'exécution du code._

2. Exécutez maintenant le code de la cellule ci-dessous.

```python
a = 6     # affectation de la valeur 6 à la variable a
b = 10    # affectation de la valeur 10 à la variable b

print('La somme de ces 2 nombres vaut :', a + b)
```

Remarquez que l'exécution du code ci-dessus n'a retourné aucune valeur. Il n'y a pas de prompt <code>Sortie[ ]</code>, uniquement l'affichage d'un texte. 

3. Peut-on **modifier le code** d'une cellule puis le **ré-exécuter** ? Bien sûr !  
Expérimentez cela dans la cellule ci-dessus en modifiant la valeur des variables <code>a</code> et <code>b</code>, puis frappez à nouveau `<maj-entrée>`
