<h2 class="alert alert-info">Exercice 3 :  $\LaTeX$ <a name="lat"></a>  </h2>

Pour que le code s'affiche, il faut le mettre entre deux dollars : <code>$ ... $</code>  
Pour écrire la formule en mode maths, centrée sur une ligne, il faut doubler les dollars : <code>$$ ... $$</code> voir [fractions](#frac) pour comprendre.  
On trouvera une liste plus complète des commandes Latex en cliquant <a href="https://fr.wikibooks.org/wiki/LaTeX/%C3%89crire_des_math%C3%A9matiques ">ici</a> mais ce qui suit devrait largement suffire!
  
## 1. Bases <a name="bas"></a>

### a. multiplication <a name="mul"></a> 
On utilise <code>\times</code> pour obtenir le symbole « multiplié par » $\times$.  
Exemple : <code>2\times3=6</code> donne $2\times3=6$. 
### b. racines <a name="rac"></a>  
<code>\sqrt{a+b}</code> pour écrire $\sqrt{a+b}$, mais on peut se contenter de <code>\sqrt a</code> pour écrire $\sqrt a$. 

Pour les racines nième, on écrira <code>\sqrt[n]{}</code>. Par exemple, <code>\sqrt[3]{8}</code> pour $\sqrt[3]{8}$.

### c. puissances<a name="pui"></a>  
<code>x^{2n-3}</code> pour écrire $x^{2n-3}$ , mais on peut se contenter de <code>x^n</code> pour écrire $x^n$.  


### d. fractions <a name="frac"></a>
On utilise <code>\frac{}{}</code> pour obtenir le trait de fraction.  
Exemple : on peut taper <code>\frac{x+2}{2x-3}</code> pour écrire $\frac{x+2}{2x-3}$, mais on peut se contenter de  <code>\frac ab </code>pour écrire $\frac ab $.  
<b>Remarques :</b> on pourra utiliser <code>\tfrac</code> pour avoir des fractions au sein du texte qui ne modifient pas l'interligne (exemple : un « petit » $\tfrac{1}{2}$), et de <code>\dfrac</code> pour avoir des grosses fractions.
Il est d'ailleurs parfois utile de combiner les deux :
$$\dfrac{1 + \tfrac{1}{2}}{3} \text{ est plus joli que } \dfrac{1 + \dfrac{1}{2}}{3}$$
<code>\frac</code> a l'avantage de s'adapter au contexte : il vaut <code>\tfrac</code> dans le corps du texte comme ici $\frac{0}{0}$ et <code>\dfrac</code> en mode maths centré : $$\pi \approx \frac{22}{7}$$.

### e. parenthèses de bonne taille <a name="par"></a>
Écrire <code>\left(</code> et <code>\right)</code> pour que la parenthèse prenne la bonne taille.  
Exemple : <code>5(2 +\frac{\sqrt x}{2})</code> donne $5(2 +\frac{\sqrt x}{2})$, tandis que <code>5\left(2 +\frac{\sqrt x}{2}\right)</code> donne $5\left(2 +\frac{\sqrt x}{2}\right)$  
<code>\left</code> et <code>\right</code> s'applique à toutes sortes de parenthèses : <code>\left[</code> et <code>\right]</code>, <code>\left\{</code> et <code>\right\}</code>, <code>\left|</code> ou <code>\left\vert</code> pour les valeurs absolues |, <code>\left\Vert</code> ou <code>\left\\|</code> pour les normes $\left\|u\right\|$.

<h2 class="alert alert-info">Exercice 5 :  insérer une animation GeoGebra   </h2>

Pour insérer une animation GeoGebra, il suffit de recopier dans **une cellule de code** ce qui suit en remplaçant *code iframe* par le code iframe du fichier GeoGebra.  
`from IPython.display import HTML
HTML("""code iframe""")`  
Vous avez un exemple ci-dessous.

```python
from IPython.display import HTML
HTML("""<iframe src="https://www.geogebra.org/classic/xwvfx23y?embed" width="800" height="600" allowfullscreen style="border: 1px solid #e4e4e4;border-radius: 4px;" frameborder="0"></iframe>""")
```

<h2 class="alert alert-info">Compléments :  raccourcis les plus utiles sur Capytale    </h2>

On retrouve tous les raccourcis dans le menu « Aide », mais il est bon de retenir ce qui suit.  <a name="cop"></a> 

Il faut d'abord cliquer sur la gauche de la cellule (partie blanche) puis, pour  
**1. Copier :** taper <code>c</code>   
Couper : <code>x</code>  
Coller : <code>v</code>  
Supprimer :  <code>d,d</code> (c'est-à-dire taper deux fois sur d)  


**2. Insérer une cellule avant :** <code>a</code> (comme above) <a name="ins"></a>  
Insérer une cellule après : <code>b</code>(comme below)

Pour mettre la cellule en Mardown : <code>m</code>  
Pour mettre la cellule en code : <code>y</code>

**4. Pour exécuter une cellule :** <code>shift + enter</code>   <a name="exe"></a>
