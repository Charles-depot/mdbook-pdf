<h2 class="alert alert-info">Exercice 2 : code IATA</h2>

Les [numéros de vol](https://fr.wikipedia.org/wiki/Num%C3%A9ro_de_vol) qui apparaissent sur les billets d'avion sont composés :
- du [code IATA](https://fr.wikipedia.org/wiki/Liste_des_codes_AITA_des_compagnies_a%C3%A9riennes) de la compagnie aérienne, **sur deux caractères**,
- d'un espace,
- d'un nombre comportant entre 1 et 4 chiffres (de 1 à 9999) déterminé par la compagnie pour désigner le vol en question.

Par exemple :
- `AF 0740` désigne un vol de la compagnie Air-France entre [Paris et Pointe-à-Pitre](https://info.flightmapper.net/fr/flight/Air_France_AF_740) : le code IATA de Air-France est donc `'AF'`.
- `LX 160` désigne un vol de la compagnie Swiss International Air Lines entre [Zurich et Tokyo](https://info.flightmapper.net/fr/flight/SWISS_LX_160) : le code IATA de Swiss International Air Lines est donc `'LX'`.


1. Exécuter les cellules ci-dessous :

````python
def code_iata(numero):
    pl = numero[0]
    dl = numero[1]
    return pl + dl
````

````python
code_iata("AF 0740")
````

````python
code_iata("LX 160")
````

````python
code_iata("UA 1")
````

````python
code_iata("EK 2") == "EK"
````

````python
code_iata("AA 45") == "AA"
````
